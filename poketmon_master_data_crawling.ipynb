```python
import requests
from bs4 import BeautifulSoup
```

```python
html = requests.get('http://web.archive.org/web/20140301191716/http://pokemondb.net/pokedex/national')
```

```python
html = BeautifulSoup(html.text)
```

```python
div = html.find('div', class_ = 'infocard-tall-list').find_all('span')
```

```python
print(div)
```

```python
list = []
for span in div:
    id = int(span.find('small').text[1:])
    name = span.find('a', 'ent-name').text
    type_list = span.find('small', class_ = 'aside').find_all('a')
    type_ = []
    for a in type_list:
        type_.append(a.text)
    list.append([id, name, type_])
```
